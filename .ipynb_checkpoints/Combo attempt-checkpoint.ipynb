```python
import sys
import sklearn
import matplotlib
matplotlib.use('Qt5Agg')
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import kneed
from kneed import KneeLocator
```

```python
inertias = []
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
data = list(zip(x, y))
labels = ["dog", "dog", "cat", "dog", "dog", "cat", "cat", "dog", "cat", "cat"]
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
```

```python
%matplotlib inline
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertiaidan')
plt.show()
```

```python
kl = KneeLocator(range(1, 11), inertias, curve="convex", direction="decreasing")
```

```python
kmeans = KMeans(n_clusters=kl.elbow)
kmeans.fit(data)
```

```python
%matplotlib inline
plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```

```python
def countN(n, nlist):
    q=0
    for x in range(len(nlist)):
        if nlist[x]==n:
            q+=1
    return q
```

```python
aMatch=0
bMatch=0
cMatch=0
dMatch=0
a=countN(1, kmeans.labels_)
b=countN(0, kmeans.labels_)

        
for x in range(len(labels)):
    if (labels[x]=="cat") and (kmeans.labels_[x]==1):
        aMatch+=1
    elif (labels[x]=="dog") and (kmeans.labels_[x]==1):
        bMatch+=1
    elif (labels[x]=="dog") and (kmeans.labels_[x]==0):
        cMatch+=1
    else:
        dMatch+=1

if aMatch>bMatch:
    print("1 represents cluster cat and 0 represents cluster dog")
    print("accuracy of cluster cat: " + str((aMatch/a)*100) +"%")
    print("accuracy of cluster dog: " + str((cMatch/b)*100) +"%")
else:
    print("1 represents cluster b and 0 represents cluster a")
    print("accuracy of cluster cat: " + str((dMatch/b)*100) +"%")
    print("accuracy of cluster dog: " + str((cMatch/a)*100) +"%")
```
